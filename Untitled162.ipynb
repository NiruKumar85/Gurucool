```python
# Handle missing values: For numeric columns, fill missing values with 0 (if appropriate), or median for non-zero distributions.
numeric_columns = ['chatSeconds', 'timeDuration', 'amount', 'astrologerOnCallDuration', 
                   'astrologersEarnings', 'netAmount', 'userOnCallDuration']
data[numeric_columns] = data[numeric_columns].fillna(0)
```

```python
# Summary statistics for key numeric columns
summary_stats = data[numeric_columns].describe()
```

```python
# Visualize the distribution of call charges using a histogram
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
data['amount'].hist(bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Call Charges', fontsize=16)
plt.xlabel('Call Charges (Amount)', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.grid(False)
plt.show()
```

```python
# Task 2: Call Center Performance Metrics

# 5. Average TalkTime for different call activities
average_talktime_by_activity = data.groupby('consultationType')['timeDuration'].mean()
```

```python
# 6. Most common source of calls
most_common_source = data['source'].value_counts().idxmax()
```

```python
# 7. Total earnings and spending for users and masters
total_earnings_users = data['amount'].sum()
total_earnings_masters = data['astrologersEarnings'].sum()
```

```python
# 8. Relationship between TalkTime and Charge (Correlation)
talktime_charge_correlation = data[['timeDuration', 'amount']].corr()
```

```python
average_talktime_by_activity, most_common_source, total_earnings_users, total_earnings_masters, talktime_charge_correlation
```

```python
# Task 3: Call Handling Analysis

# Calculate additional columns for call handling analysis
data['ConnectTime'] = pd.to_numeric(data.get('astrologerOnCallDuration', 0), errors='coerce')
data['DialTime'] = pd.to_numeric(data.get('userOnCallDuration', 0), errors='coerce')
data['UnconnectTime'] = pd.to_numeric(data.get('timeDuration', 0), errors='coerce')
```

```python
# 9. Average time for calls to be connected (ConnectTime - DialTime)
data['ConnectionDelay'] = data['ConnectTime'] - data['DialTime']
average_connection_time = data['ConnectionDelay'].mean()
```

```python
# 10. Most common reason for call disconnection (UnconnectTime - ConnectTime)
data['DisconnectionDelay'] = data['UnconnectTime'] - data['ConnectTime']
most_common_disconnection_reason = data['DisconnectionDelay'].value_counts().idxmax()
```

```python
# 11. Analyze HangUpTime patterns (if relevant)
hangup_time_pattern = data['userOnCallDuration'].describe()
```

```python
average_connection_time, most_common_disconnection_reason, hangup_time_pattern
```

```python
# Task 4: Order and Refund Analysis

# 12. Determine the order status distribution for calls
order_status_distribution = data['callStatus'].value_counts()
```

```python
# 13. Calculate the total refund amount and identify the refund status distribution
total_refund_amount = data['refundAmount'].sum() if 'refundAmount' in data else 0
refund_status_distribution = data['refundStatus'].value_counts() if 'refundStatus' in data else None
```

```python
order_status_distribution, total_refund_amount, refund_status_distribution
```

```python
# Task 5: Insights and Recommendations

# Insights: Summarizing key performance indicators
insights = {
    "most_common_source": most_common_source,
    "total_user_spending": total_earnings_users,
    "total_master_earnings": total_earnings_masters,
    "talktime_charge_correlation": talktime_charge_correlation.loc['timeDuration', 'amount'],
    "average_talktime_by_activity": average_talktime_by_activity.to_dict(),
    "order_status_distribution": order_status_distribution.to_dict() if 'order_status_distribution' in locals() else None,
    "refund_status_distribution": refund_status_distribution.to_dict() if refund_status_distribution is not None else None,
    "total_refund_amount": total_refund_amount if 'total_refund_amount' in locals() else 0,
}
```

```python
# Display insights and recommendations
print("### Insights ###")
for key, value in insights.items():
    print(f"{key}: {value}")
```
